*Importing Libraries*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

*Importing CSV File*

```python
file_path = 'CrashReport.csv'
data = pd.read_csv(file_path, index_col = 0)
```

*Exploratory Data Analysis*

```python
data.head()
```

```python
data.describe()
```

```python
print(data.isnull().sum())
```

```python
data.replace('-', np.nan, inplace = True)
data = data.dropna()
```

```python
data.describe()
```

*Renaming indexes for cleaning purposes*

```python
data = data.rename(index = {'&nbsp;&nbsp;Motorcyclists' : 'Motorcyclists'})
```

```python
data = data.rename(index = {'Unknown' : 'Unknown1', 'Other/ Unknown' : 'Unknown2', 'Total*' : 'Total'})
```

*Transposing Database to flip **Rows** with **Cloumns***

```python
data = data.transpose()
```

```python
data = data.drop(data.index[-1])
```

*Saving the modified CSV into a new CSV file*

```python
data.to_csv('modified_CrashReport.csv', index = True)
```
